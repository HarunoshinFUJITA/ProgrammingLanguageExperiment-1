```python
'''演習1-1: ファイルの取り扱い'''

f = open('fread1.ipynb', 'r') # f = open('README.md', 'r') なども試す
lines = f.readlines() # ファイル全部を1つの文字列に読み込む
f.close()

# print(lines)

# data1 に文字列として読み込めているのが確認できれば，
# あとは文字列の処理を行っていく．
# 以下に演習の答えを書いていくこと
line_count = 1;
for l in lines:
    l = l.strip()
    print('%03d: ' % (line_count), l)
    line_count += 1
```
